```python
import testspace_colab
trove = testspace_colab.Trove() # Create the trove instance. defaults to ~/testspace-colab
```

We load all the results from the trove and display them in data frames

```python
from IPython.display import display
import jsonpath_ng
import pandas as pd
max_load=3
for result in trove.load(): # this is a generator - we only want 2
    print(f"NAME={result['name']} ID={result['id']}")
    df = pd.DataFrame([match.value for match in jsonpath_ng.parse('$..[cases][:]').find(result)])
    display(df)
    max_load-=1
    if max_load==0:
        break
```
