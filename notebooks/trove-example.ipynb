# Trove Example
This illustrates how to load results from the trove

```python
import testspace_colab.trove as trove_module
trove = trove_module.Trove()
```

```python
from IPython.display import display
import jsonpath_ng
import pandas as pd
for result in trove.load():
    df = pd.DataFrame([match.value for match in jsonpath_ng.parse('$..[cases][:]').find(result)])
    display(df)
```
