# Initialization

This is a very simple example using the API. We load a default configuration pointing to 
https://lbrack.testspace.com/projects/samples/spaces/main using ``testspace_colab.use_test_config()`` and initialize the API

```python
import testspace_colab
testspace_colab.use_test_config()
client = testspace_colab.API()
```

# Some magic

The following code snippet creates a list of test cases from the response but running a json-path query
similar to xpath. 

```python
import jsonpath_ng
import pandas as pd
frame_result_details = dict()
# We load all the test cases as pandasDataFrames
for result in client.get_results( project='lbrack:testspace-colab', space='flat-structure' ):
    print(result['id'], result['name'])
    frame_result_details[result['name']] = pd.DataFrame([match.value for match in jsonpath_ng.parse('$..[cases][:]').find( client.get_result_details(result=result['id'], project='lbrack:testspace-colab', space='flat-structure'))])
```

```python
from IPython.display import display
import xarray

xr = xarray.Dataset(frame_result_details)
display(xr)
```
